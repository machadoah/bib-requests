# Conhecendo a requests

## Primeira requisição

Uma requisição é uma comunicação de um cliente para um servidor, podendo ser solicitações, envios, modificações e deleções de dados. 

**Status codes:**

- 2xx: Sucesso!
- 3xx: Redirecionamento! Deve-se realizar mais operações para concluir a operação
- 4xx: Erro do cliente! 
- 5xx: Erro do servidor!

```python
import requests
```

```python
r = requests.get('https://api.github.com/events')
```

```python
r
```

Podemos ver que a requisição acima foi aceita e **executada com sucesso!**
